```python
from fitparse import FitFile
import pandas as pd
import glob

fitfiles = glob.glob("'/Users/Isaac/Downloads/export_34470729/activities/*.gz")

fitfile = FitFile('/Users/Isaac/Downloads/export_34470729/activities/1948436385.fit')

fitdf = pd.DataFrame()

# Get all data messages that are of type record
for record in fitfile.get_messages('record'):
    r={}
    # Go through all the data entries in this record
    for record_data in record:
        # Print the records name and value (and units if it has any)
        #if record_data.units:
            # print (" * %s: %s %s" % (
            #    record_data.name, record_data.value, record_data.units,
            # ))
        #else:
            # print (" * %s: %s" % (record_data.name, record_data.value))
        r[record_data.name]=record_data.value
    fitdf = pd.concat([fitdf,pd.DataFrame(r,index=[0])],ignore_index=True,sort=False)
fitdf
```

```python
import glob
import gzip
import io

allfitdf = pd.DataFrame()

fitfiles = glob.glob("/Users/Isaac/Downloads/export_34470729/activities/*.gz")

for file in fitfiles:
    print(file)
    with gzip.open(file, 'rb') as f:
        content = f.read()

        fitfile = FitFile(io.BytesIO(content))

        # Get all data messages that are of type record
        for record in fitfile.get_messages('record'):
            r={}
            r["filename"]=file
            # Go through all the data entries in this record
            for record_data in record:
                r[record_data.name]=record_data.value
            fitdf = pd.concat([fitdf,pd.DataFrame(r,index=[0])],ignore_index=True,sort=False)
    allfitdf = pd.concat([allfitdf,fitdf],ignore_index=True,sort=False)
```

```python
allfitdf.to_csv("AllFITdf.csv",index=False)
```
